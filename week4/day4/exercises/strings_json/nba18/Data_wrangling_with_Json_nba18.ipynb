## Working on Jsons

```python
import requests
import json
import pandas as pd
```

```python
r = requests.get(url='http://data.nba.net/prod/v2/2018/teams.json')
json_readed = json.loads(r.text)
json_readed
```

```python
print(json_readed)
```

Save json in a local file called "data.json"

```python
with open('data.json', 'w+') as outfile:
    json.dump(json_readed, outfile)
```

```python
with open('data.json') as f:
    line = f.readline()
    if line:
        print(line)
        line = f.readline()
print(line)
```

Save with indent

```python
with open('data_indented.json', 'w+') as outfile:
    json.dump(json_readed, outfile, indent=4)
```

Read local json


```python
with open('data_indented.json', 'r+') as outfile:
    json_readed = json.load(outfile)
json_readed
```

```python
df_json = pd.read_json("data_indented.json")
df_json
```

### Data Mining & Data Wrangling
As you can see, there are jsons inside the original json. For that, we have to mining the json (data mining) to be able to use it correctly (data wrangling). 

How do you solve this issue? Research about this and try a solution. 

```python
with open('data_indented.json', 'r') as outfile:
    json_readed = json.load(outfile)
    for dict_ in json_readed["league"]:
        print(dict_)
        break
```

```python
with open("data_indented.json", "r+") as outfile:
    json_readed = json.load(outfile)
    standard_json_list = json_readed["league"]["standard"] 
# standard_json_list
```

```python
with open('final_standard.json', 'w+') as outfile:
    json.dump(standard_json_list, outfile, indent=4)
```

```python
with open("data_indented.json", "r+") as outfile:
    json_readed = json.load(outfile)
    africa_json_list = json_readed["league"]["africa"] 
with open('final_africa.json', 'w+') as outfile:
    json.dump(africa_json_list, outfile, indent=4)
```

```python
with open("data_indented.json", "r+") as outfile:
    json_readed = json.load(outfile)
    sacramento_json_list = json_readed["league"]["sacramento"] 
with open('final_sacramento.json', 'w+') as outfile:
    json.dump(sacramento_json_list, outfile, indent=4)
```

```python
with open("data_indented.json", "r+") as outfile:
    json_readed = json.load(outfile)
    vegas_json_list = json_readed["league"]["vegas"] 
with open('final_vegas.json', 'w+') as outfile:
    json.dump(vegas_json_list, outfile, indent=4)
```

```python
with open("data_indented.json", "r+") as outfile:
    json_readed = json.load(outfile)
    utah_json_list = json_readed["league"]["utah"] 
with open('final_utah.json', 'w+') as outfile:
    json.dump(utah_json_list, outfile, indent=4)
```

```python
df_standard = pd.read_json("final_standard.json")
df_africa = pd.read_json("final_africa.json")
df_sacramento = pd.read_json("final_sacramento.json")
df_vegas = pd.read_json("final_vegas.json")
df_utah = pd.read_json("final_utah.json")
```

```python
final_df = pd.concat([df_standard, df_africa, df_sacramento, df_vegas, df_utah], axis=0, sort=True)
final_df
```
